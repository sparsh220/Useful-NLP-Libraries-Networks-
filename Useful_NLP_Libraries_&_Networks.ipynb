## Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.

Answer: NLTK and spaCy are both Python libraries used for Natural Language Processing (NLP). They help computers understand human language. However, they are different in purpose, ease of use, and speed.

NLTK (Natural Language Toolkit) is one of the oldest NLP libraries. It is mainly used for learning and research. It provides many small tools for tasks like tokenization, stemming, lemmatization, POS tagging, and parsing. It also has strong linguistic resources like WordNet and many text corpora. Because of this, NLTK is very useful for students and beginners who want to understand how NLP works step by step. However, NLTK is slower and often needs more code to perform tasks. It is not designed mainly for large or real-world applications.

spaCy is a newer and more modern NLP library. It is designed for building real applications. spaCy provides fast and ready-to-use models for tasks like tokenization, part-of-speech tagging, named entity recognition, and dependency parsing. Its code is clean and simple, so it is easier to use for practical work. spaCy also supports deep learning models and works well with large amounts of text. Because of this, spaCy is much faster than NLTK and is commonly used in production systems.

In terms of features, NLTK focuses more on linguistic tools and teaching resources, while spaCy focuses on modern NLP models and real-world usage. In terms of ease of use, spaCy is easier and more direct, while NLTK can feel complex and needs more steps. In terms of performance, spaCy is faster and more efficient, whereas NLTK is slower but good for learning.

## Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?

Answer: TextBlob is a simple Python library used for Natural Language Processing (NLP). It is built on top of other NLP tools and makes text processing easy for beginners. TextBlob is mainly used for basic tasks like sentiment analysis, translation, tokenization, and part-of-speech tagging. It allows users to work with text using very simple and readable code.

TextBlob simplifies sentiment analysis by giving a direct result about the feeling of a sentence. It tells whether the text is positive, negative, or neutral. It also gives two values:

* Polarity (how positive or negative the text is)

* Subjectivity (how much opinion is in the text)
This means users do not need to train complex models. They can just pass a sentence, and TextBlob returns the sentiment easily.

TextBlob also simplifies translation by using built-in translation functions. A user can write one line of code to translate text from one language to another. For example, it can translate English text into Hindi or French without writing complex translation logic. This makes it very useful for beginners who want quick results.

In addition to this, TextBlob supports tasks like noun phrase extraction, spelling correction, and text classification. Because of its simple syntax and easy setup, TextBlob is widely used for small projects and learning purposes.

## Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).

Answer: A Recurrent Neural Network (RNN) is a type of neural network used to process sequence data, such as text, speech, or time-series data. Unlike normal neural networks, an RNN has a memory that helps it remember past information while processing new input. This makes it useful for tasks like language translation, speech recognition, and sentence prediction.

Architecture of RNN

The architecture of an RNN has three main parts:

* Input layer – takes data one step at a time (for example, one word at a time).

* Hidden layer – stores information from previous steps. It connects to itself, which is why it is called “recurrent.”

* Output layer – gives the result for each step or at the end of the sequence.

At each time step, the RNN takes the current input and also the previous hidden state (memory). It combines them to produce a new hidden state. This hidden state is then used to predict the output and is passed to the next step.

Functioning of RNN

* When an RNN reads a sentence, it does not read all words at once. It reads them one by one.

* First, it reads the first word and saves some information about it.

* Then it reads the second word and uses both the new word and the saved memory from the first word.

This process continues until the end of the sentence.

Because of this, an RNN can understand the order of words and the context of a sentence. However, simple RNNs have a problem remembering very long sequences. This problem is called the vanishing gradient problem. To solve this, improved models like LSTM and GRU are used.

## Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

Answer: The key difference between LSTM and GRU networks is in their structure and gates.

LSTM (Long Short-Term Memory) has three gates:

* Input gate – decides what new information to store

* Forget gate – decides what information to remove

* Output gate – decides what information to send as output

Because of these gates, LSTM can remember information for a long time, but it is more complex and uses more parameters.

GRU (Gated Recurrent Unit) has two gates:

* Update gate

* Reset gate

GRU combines the cell state and hidden state into one, so its design is simpler than LSTM. It usually trains faster and needs less memory.

In NLP applications:

* LSTM is good when very long memory is needed.

* GRU is good when speed and simplicity are more important.

```python
# Question 6: Write a Python program using TextBlob to perform sentiment analysis on
# the following paragraph of text:
# “I had a great experience using the new mobile banking app. The interface is intuitive,
# and customer support was quick to resolve my issue. However, the app did crash once
# during a transaction, which was frustrating"
# Your program should print out the polarity and subjectivity scores.
# (Include your Python code and output in the code box below.)


from textblob import TextBlob

text = "I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating."

blob = TextBlob(text)

polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print("Polarity:", polarity)
print("Subjectivity:", subjectivity)
```

```python
# Question 7: Given the sample paragraph below, perform string tokenization and
# frequency distribution using Python and NLTK:
# “Natural Language Processing (NLP) is a fascinating field that combines linguistics,
# computer science, and artificial intelligence. It enables machines to understand,
# interpret, and generate human language. Applications of NLP include chatbots,
# sentiment analysis, and machine translation. As technology advances, the role of NLP
# in modern solutions is becoming increasingly critical.”
# (Include your Python code and output in the code box below.)

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

tokens = word_tokenize(text.lower())

freq_dist = FreqDist(tokens)

print("Tokens:")
print(tokens)

print("\nFrequency Distribution:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```

```python
# Question 8: Implement a basic LSTM model in Keras for a text classification task using
# the following dummy dataset. Your model should classify sentences as either positive
# (1) or negative (0).
# Dataset
# texts = [
# “I love this project”, #Positive
# “This is an amazing experience”, #Positive
# “I hate waiting in line”, #Negative
# “This is the worst service”, #Negative
# “Absolutely fantastic!” #Positive]
# labels = [1, 1, 0, 0, 1]
# Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
# this data. You may use Keras with TensorFlow backend.
# (Include your Python code and output in the code box below.)



import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic"
]
labels = [1, 1, 0, 0, 1]

# Tokenization
tokenizer = Tokenizer(num_words=50)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding
padded_sequences = pad_sequences(sequences, maxlen=6)

# Convert labels to array
labels = np.array(labels)

# Build LSTM model
model = Sequential()
model.add(Embedding(input_dim=50, output_dim=16, input_length=6))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(padded_sequences, labels, epochs=10, verbose=1)

# Test on same data
predictions = model.predict(padded_sequences)
print("Predictions:")
print(predictions)
```

```python

# Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
# lemmatization, and entity recognition. Use the following paragraph as your dataset:
# “Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
# development of India’s atomic energy program. He was the founding director of the Tata
# Institute of Fundamental Research (TIFR) and was instrumental in establishing the
# Atomic Energy Commission of India.”
# Write a Python program that processes this text using spaCy, then prints tokens, their
# lemmas, and any named entities found.
# (Include your Python code and output in the code box below.)


import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# Process text
doc = nlp(text)

print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)

print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# Question 10: You are working on a chatbot for a mental health platform. Explain how
# you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
# NLP to understand and respond to user input effectively. Detail your architecture, data
# preprocessing pipeline, and any ethical considerations.
# (Include your Python code and output in the code box below.)

import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

texts = ["I feel very sad", "I am happy today", "I feel stressed"]
labels = [0, 1, 0]  # 0 = negative, 1 = positive

# Preprocess using spaCy
clean_texts = []
for t in texts:
    doc = nlp(t.lower())
    clean_texts.append(" ".join([token.lemma_ for token in doc]))

# Tokenize
tokenizer = Tokenizer(num_words=20)
tokenizer.fit_on_texts(clean_texts)
sequences = tokenizer.texts_to_sequences(clean_texts)
padded = pad_sequences(sequences, maxlen=5)

labels = np.array(labels)

# Build LSTM model
model = Sequential()
model.add(Embedding(20, 8, input_length=5))
model.add(LSTM(8))
model.add(Dense(1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(padded, labels, epochs=5, verbose=1)

# Test
test_text = ["I feel very happy"]
doc = nlp(test_text[0].lower())
test_clean = [" ".join([token.lemma_ for token in doc])]
test_seq = tokenizer.texts_to_sequences(test_clean)
test_pad = pad_sequences(test_seq, maxlen=5)

pred = model.predict(test_pad)
print("Prediction:", pred)
```
